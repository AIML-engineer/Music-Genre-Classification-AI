```python
# import libraries
import pandas as pd
```

```python
train_dataset = pd.read_csv('/content/drive/Shareddrives/1:1 Srinithi Rajan/Dataset/Data/CSV files/train_scaled_3_sec.csv')
```

```python
# Train a Random forest algorithm with 5-fold cross-validation
X = train_dataset.drop(['label'], axis =1)
y = train_dataset['label']
```

```python
# import cross_val_score and Random forest
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import RandomForestClassifier
import numpy as np

# Random Forest
random_forest = RandomForestClassifier(n_estimators=100, max_depth=20, random_state=0)

# Perform 5-fold cross-validation
cv_scores = cross_val_score(random_forest, X, y, cv=5)

# Print the cross-validation scores for each fold
print(cv_scores)

# The average can be considered as the overall performance of the model
print("Average 5-Fold CV Score: ", np.mean(cv_scores))
# Please the average accuracy value into this sheet - https://docs.google.com/spreadsheets/d/1ly0Mgfquvcxa-1lU7bWTDoDEvc6t3ZZRrOL3DNBCtl8/edit?usp=sharing
```

```python
# Create a Random forest classifier
import pickle
rf = RandomForestClassifier(n_estimators=100, max_depth=20, random_state=0)
rf.fit(X,y)
with open("/content/drive/Shareddrives/1:1 Srinithi Rajan/Model/rf_model_3sec.pkl","wb") as file:
  pickle.dump(rf,file)
  file.close()
```
