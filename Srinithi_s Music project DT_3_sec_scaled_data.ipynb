```python
# import libraries
import pandas as pd
```

```python
train_dataset = pd.read_csv('/content/drive/Shareddrives/1:1_Srinithi_Rajan/Dataset/Data/CSV files/train_scaled_3_sec.csv')
```

```python
# Train a Decision tree algorithm with 5-fold cross-validation
X = train_dataset.drop(['label'], axis =1)
y = train_dataset['label']
```

```python
# import cross_val_score and Decision tree
from sklearn.model_selection import cross_val_score
from sklearn.tree import DecisionTreeClassifier
import numpy as np

# Decision trees
decision_tree = DecisionTreeClassifier()

# Perform 5-fold cross-validation
cv_scores = cross_val_score(decision_tree, X, y, cv=5)

# Print the cross-validation scores for each fold
print(cv_scores)

# The average can be considered as the overall performance of the model
print("Average 5-Fold CV Score: ", np.mean(cv_scores))
# Please the average accuracy value into this sheet - https://docs.google.com/spreadsheets/d/1ly0Mgfquvcxa-1lU7bWTDoDEvc6t3ZZRrOL3DNBCtl8/edit?usp=sharing
```

```python
# Create a Decision tree classifier
import pickle
dt = DecisionTreeClassifier()
dt.fit(X,y)
with open("/content/drive/Shareddrives/1:1_Srinithi_Rajan/Model/dt_model_3sec.pkl","wb") as file:
  pickle.dump(dt,file)
  file.close()
```
