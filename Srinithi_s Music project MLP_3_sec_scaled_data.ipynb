```python
# import libraries
import pandas as pd
```

```python
train_dataset = pd.read_csv('/content/drive/Shareddrives/1:1 Srinithi Rajan/Dataset/Data/CSV files/train_scaled_3_sec.csv')
```

```python
# Train a Random forest algorithm with 5-fold cross-validation
X = train_dataset.drop(['label'], axis =1)
y = train_dataset['label']
```

```python
# import cross_val_score and Random forest
from sklearn.model_selection import cross_val_score
from sklearn.neural_network import MLPClassifier
import numpy as np

# MLP with learning rate and epochs
mlp = MLPClassifier(
    solver='adam',
    alpha=1e-5,
    hidden_layer_sizes=(5000, 10),
    learning_rate_init=0.005,  # Learning rate
    max_iter=100,  # Number of epochs
    random_state=1
)

# Perform 5-fold cross-validation
cv_scores = cross_val_score(mlp, X, y, cv=5)

# Print the cross-validation scores for each fold
print(cv_scores)

# The average can be considered as the overall performance of the model
print("Average 5-Fold CV Score: ", np.mean(cv_scores))
```

```python
# Create a Random forest classifier
import pickle
mlp = MLPClassifier(
    solver='adam',
    alpha=1e-5,
    hidden_layer_sizes=(5000, 10),
    learning_rate_init=0.001,  # Learning rate
    max_iter=10,  # Number of epochs
    random_state=1
)
mlp.fit(X,y)
with open("/content/drive/Shareddrives/1:1 Srinithi Rajan/Model/mlp_model_3sec_0.001_10.pkl","wb") as file:
  pickle.dump(mlp,file)
  file.close()
```
